```python
# install panda: !pip install pandas
# A Series is essentially a column, and a DataFrame is a multi-dimensional table made up of a collection of Series.

import pandas as pd
# series
x=[10,20,30,40]
series=pd.Series(x)
# change index name
series2=pd.Series(x, index=['a','b','c','d'])
print(series)
print(series2)
```

```python
a=pd.Series(12)
print(a)
b=pd.Series(10, index=[1,2,3,4])
print(b)
```

```python
#dataframe
data={
    'name':['alex','david','robert', 'jack', 'molina'], 'rank':[2,5,3,1,7], 'marks':[90,34,53,25,69]
}
# create a dataframe and set the index name:
df=pd.DataFrame(data, index=['s1','s2','s3', 's4' ,'s5'])
print(df)

#find the colomn name:

col=df.columns
print(col)

for cols in df:
  print(cols)
```

```python
#access value in a selected row by index name:
print(df.loc['s1'])

#find  only sected colomn in a row:
print(df.loc['s2','name']) #row,col

#find column by index no:
print("the result of 2nd and 4th column is \n",df.iloc[[2,4]],)
```

```python
# dataframe attributes methods:

# find the data type
print(df.dtypes)

# dimention of the dataset
print("dimention: ",df.ndim)

print(df.shape)

# the number of elements in the dataframe:
print("total elements:", df.size)

# see the indices :
print("the idices is: ", df.index)

# transpose the table:

print("the transpose table is: \n\n",df.T)

# see the first n rows

print("\n the first two rows:\n",df.head(2)) #return first two rows

# see last n rows

print("\nthe last two rows: \n",df.tail(2))

# Getting info about your data
print(df.info())
```

```python
#merge and concat two dataset
data1={
    "id":[1,2,3,4,5],
    "name":['arif', 'asif', 'babul', 'alex', 'mina'],
     'marks':[45,53,66,77,88]
}
data2={
      "id":[4,5,8,9,10],
    "name":['alex','david','robert', 'jack', 'molina'],
     'dept':['cse','eee','it','bba','economics']
}

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)

print(pd.concat([df1,df2],axis=1))
# print(pd.merge(df1,df2,how="inner",on="id"))
print(pd.merge(df1,df2,how="right",on="id"))
```

```python
#add new colomn: join operation
data3={
    "country":['Bangladesh','Bangladesh','india','pakistan','india']
}
df3=pd.DataFrame(data3)
print(df1.join(df3))
```
